```python
def fileReader():
    namesList=[]
    with open(r"/Users/aaronSmith/Documents/p022_names.txt","r") as f:
        lines=f.readline()
        for j in lines.split(","):
            namesList.append(j.strip('""'))
      
    return namesList
```

```python
alphaString="0ABCDEFGHIJKLMNOPQRSTUVWXYZ"
namesList=fileReader()
namesList.sort()
sumList=[]
sortedPosition=1
for name in namesList:
    sumLetters=0
    for letter in name:
        sumLetters+=alphaString.find(letter)
    sumList.append(sumLetters*sortedPosition)  
    sortedPosition+=1
print(sum(sumList))
```
