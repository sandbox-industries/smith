```python
file3=file3.split(",\n")[1:]
countyList=[]
popList=[]
for index in range(len(file3)):
    countyList.append(file3[index][0:len(file3[0])].split(",")[2].replace('"',"").replace(" County",""))
    popList.append(int(file3[index][0:len(file3[index])].split(",")[3].replace('"',"")+file3[index][0:len(file3[index])].split(",")[4].replace('"',"")))
countyPopDict=dict(zip(countyList,popList))
countyPopDict
```

```python
orderedPopList=[]
for county in coloDf.county:
    if county in countyPopDict.keys():
        orderedPopList.append(countyPopDict[county])
    else:
        orderedPopList.append(np.nan)

coloDf["population"]=orderedPopList
```

```python
coloDf["infectionRate"]=coloDf.cases/coloDf.population
```
