```python
import pandas as pd
import numpy as np
```

```python
file="https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv"
```

```python
df=pd.read_csv(file)
coloDf=df[df.state=="Colorado"]
coloDf=coloDf[coloDf.county!="Unknown"]
coloDf.reset_index(inplace=True,drop=True)
```

```python
coloDf
```

```python
countyList=[]
fipsList=[]

for i in range(len(coloDf)):
    countyList.append(coloDf.iloc[i,1])
    fipsList.append("0"+str(int(coloDf.iloc[i,3])))
countyFipsDict=dict(zip(countyList,fipsList))
countyFipsDict
```

```python
coloDf=coloDf.groupby(coloDf.county).sum()
coloDf=coloDf.drop("fips",axis=1)
coloDf.reset_index(inplace=True)
```

```python
coloDf
```

```python
orderedFipsList=[]
for county in coloDf.county:
    orderedFipsList.append(countyFipsDict[county])

coloDf["fips"]=orderedFipsList
```

```python
coloDf
```

```python
file2="/Users/aaronSmith/Downloads/profile-county.csv"
df2=pd.read_csv(file2)
df2=df2.iloc[:,2:4]
df2
```

```python
countyList2=[]
popList=[]

for i in range(len(df2)):
    countyList2.append(df2.iloc[i,0].replace(" County",""))
    popList.append(int(df2.iloc[i,1].replace(",","")))
countyPopDict=dict(zip(countyList2,popList))
countyPopDict
```

```python
orderedPopList=[]
for county in coloDf.county:
    orderedPopList.append(countyPopDict[county])

coloDf["population"]=orderedPopList
```

```python
coloDf
```

```python
coloDf["infectionRate"]=coloDf.cases/coloDf.population
```

```python
coloDf.sort_values(by="infectionRate",ascending=False)
```
